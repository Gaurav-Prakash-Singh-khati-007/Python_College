```python
count=0
def linear_search(x):
    a=[10,20,30,40,89,9,876]
    for i in range(0,len(a)):
        if(a[i]==x):
            count=1
            index=a[i]
            print(f"Item found at index {i}")
    if(count==0):
        print('item not found in list ')
linear_search(30)
```
