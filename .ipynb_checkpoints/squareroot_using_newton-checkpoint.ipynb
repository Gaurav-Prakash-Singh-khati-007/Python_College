```python
# Algorithm NewtonSquareRoot(N)
# input number =a
# approximation : x=a

# if N < 0 then
#   print "Square root not defined"
#countinues loop
# epsilon=0.000001 for accuracy
# 7. repeat
# 8.     prev = x
# 9.     x = (x + N / x) / 2
# 10. until |x - prev| < epsilon

```

```python
a=int(input("Enter a number to find square root"))
if(a<0):
    print("square root cannot be achieved")
x=a
episilon=0.000001
while(True):
    prev=x
    x=(x+a/x)/2
    if(abs(x-prev)<episilon):
        break
print(x)
```
