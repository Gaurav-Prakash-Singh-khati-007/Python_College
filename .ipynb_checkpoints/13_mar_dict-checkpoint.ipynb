```python
# word frequency 
words='bronnostonsdtbronnostonsdtbronnostonsdtv'
d=dict()
for word in words:
    if(word not in d ):
        d[word]=1
    else:
        d[word]+=1
print(d)
```

```python
# get 
print(n.get('five',0))
```

```python
# word frequency  using get
words='bronnostonsdtbronnostonsdtbronnostonsdtv'
d=dict()
for word in words:
    d[word]=d.get(word,0)+1
print(d)
```
