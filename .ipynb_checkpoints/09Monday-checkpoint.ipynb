### Operation On list

```python
lst1=[1,2,3,4,56,78,87,6,5,545,4343,3456,876]
lst1.append(9889988989)
print(lst1,'append operation')
lst1.insert(0,98)
print(lst1,'insert operation')

lst1.remove(4343)
print(lst1,'remove 4343 operation')

lst1.pop()
print(lst1,'poped last elements  operation')

del lst1[2:4]
print(lst1,'del all operation')

# print(del 0,98),'insert operation')
lst1.extend([23,43,54,76,87,787,4543,23])
print(lst1,'extended operation')

print('minimum value ',min(lst1))
print('maximum value ',max(lst1))
print('sum of lst',sum(lst1))
print('sorted',lst1.sort())

lst1.clear()
print(lst1) 
```

### Copying or cloning the list

```python
a=[1,2,3,4,5,6,78]
b=a.copy()
print(b)
b.remove(4)
print(a)
print(b)
#using lst slicing
c=a[:]
print(c)
c.remove(78)
print(c)
```

```python
#list comprehension
import copy
a=[1,2,3,4,5,6,7,88,90]
b=[item for item in a]
print(b)
# nested lsit
nst_lst=[[2,3],[4,5],[6,8]]
deep_nst_lst=copy.deepcopy(nst_lst)
print(deep_nst_lst)
st=nst_lst.copy()
```

```python
# why this deep copy used 
a=[[2,3],[4,5],[6,8]]
b=a.copy()
b[0][0]=455666
print(a)
print(b)
```
