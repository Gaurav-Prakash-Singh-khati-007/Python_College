```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        key=arr[i]
        j=i-1
        while(j>=0 and key<arr[j]):
            arr[j+1]=arr[j]
            j=j-1
        arr[j+1]=key
    return arr
arr=[77,33,44,11,88,22,66,56]
result=insertion_sort(arr)
print(result)
```
