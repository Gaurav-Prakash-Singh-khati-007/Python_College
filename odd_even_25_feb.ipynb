```python
#prime no
# number only divisible by 1 and self
#loop till self

number1=int(input("enter a number "))
count=1
for i in range(2,number1):
    if(number1%i==0):
        count=count+1
if(count>=2):
    print("number is not prime number")
else:
    print("number is prime number")
```

```python
# circulating the list items
n_terms=int(input("enter the numbers want to add to list"))
list1=[]
for var in range(0,n_terms,1):
    n=int(input("enter number"))
    list1.append(n)
print(list1)
print("circulating the list")

for var in range(0,n_terms,1):
    n=list1.pop(0)
    list1.append(n)
    print(list1)
```
