```python
# selection_ sort
def min_element(arr,k):
    minimum=arr[k]
    loc=k
    for j in range(k,len(arr)):
        if(minimum>arr[j]):
            minimum=arr[j]
            loc=j
    return minimum,loc
def selection_sort(arr):
    for k in range(0,len(arr)-1):
        minimum,loc=min_element(arr,k)
        temp=arr[k]
        arr[k]=arr[loc]
        arr[loc]=temp
    return arr
arr=[77,33,44,11,88,22,66,55]
result=selection_sort(arr)
print(result)
```
