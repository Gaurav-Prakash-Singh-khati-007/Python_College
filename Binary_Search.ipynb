```python
def binary_search(arr,x):
    low=0
    high=len(arr)-1
    mid=(low+high)//2
    while(low<=high) and (arr[mid]!=x):
        if(arr[mid]<x):
            low=mid+1
        else:
            high=mid-1
        mid=(low+high)//2
    if(arr[mid]==x):
        return mid
    else:
        return -1
arr=[1,2,3,4,5,6,7,8,9]
x=9
result=binary_search(arr,x)
if(result!=-1):
    print("item found at index",result)
else:
    print("Item not found at any where")

            
```
