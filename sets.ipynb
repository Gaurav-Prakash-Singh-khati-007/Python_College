# join operation in set 


```python
s1={1,2,3}
s2={4,5,6}
# s3=s1.union(s2)
s3=s1|s2
s3
```

```python
## intersection
s1={1,2,3,4,5}
s2={1,3,5}
# s3=s1.intersection(s2)

s3=s1&s2
s3
```

```python
# intersection_update
s1={1,2,3}
s2={3,4,5}
s1.intersection_update(s2)
s1
```

```python
# diffrence operation
s1={1,2,3}
s2={3,4,5}
# s3=s1.difference(s2)
s3=s1-s2
s3
```

```python
#symmetric diffrence
s1={1,2,3}
s2={3,4,5}
# s3=s1.symmetric_difference(s2)
# s3=s1^s2
s1.symmetric_difference_update(s2)
s1
```
