### Passing list as a function   

```python
def fruits(fru):
    for fruit in fru:
        print(fruit,end=' ') # end='' used as end then  at end of once print 
fruit=['Mango','Apple','Grapes']
fruits(fruit)
```

## list loops

```python
# list loop using indexing
thislist=['apple','banana','grapes']
for i in range(len(thislist)):
    print(thislist[i])
```

```python
# while loop
thislist=['apple','banana','grapes']
i=0

while(i<len(thislist)):
    print(thislist[i])
    i+=1
```

```python
# while loop
thislist=['apple','banana','grapes']
i=0

while(i<len(thislist)):
    print(thislist[i])
    i=i+1
```

```python
# looping using list comprehension
thislist=['apple','banana','grapes']
[print(x) for x in thislist ]
```
